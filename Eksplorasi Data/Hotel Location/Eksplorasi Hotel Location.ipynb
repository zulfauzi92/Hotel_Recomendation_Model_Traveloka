### Merge Data 

```python
# get latitude and longitude value for each city
for i,item in df_city_location.iterrows():
    df_city_location.loc[i, 'latitude'] = (df_us_city.loc[df_us_city['city'] == item['city'], 'lat'].mean())
    df_city_location.loc[i, 'longitude'] = (df_us_city.loc[df_us_city['city'] == item['city'], 'lng'].mean())
    
# drop null value
df_city_location.dropna(inplace=True)

# rename city to location_name
df_city_location.rename(columns={'city':'location_name'}, inplace=True)
len(df_city_location)
```

```python
# get latitude and longitude value for each state
df_state_location = pd.DataFrame(df_us_city['state_id'].unique())
df_state_location.rename(columns={0:'state'}, inplace=True)
for i,item in df_state_location.iterrows():
    df_state_location.loc[i, 'state_name'] = (df_us_city.loc[df_us_city['state_id'] == item['state'], 'state_name'].mode().values[0])
    df_state_location.loc[i, 'latitude'] = (df_us_city.loc[df_us_city['state_id'] == item['state'], 'lat'].mean())
    df_state_location.loc[i, 'longitude'] = (df_us_city.loc[df_us_city['state_id'] == item['state'], 'lng'].mean())

df_state_location.drop(columns=['state'], inplace=True)
df_state_location.rename(columns={'state_name':'location_name'}, inplace=True)
len(df_state_location)
```

```python
# merge city and state as 1 dataframe
df_location = pd.merge(df_state_location, df_city_location, how='outer')
df_location
```
